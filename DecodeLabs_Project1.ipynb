```python
# ==============================
# DecodeLabs Project 1
# Data Cleaning & Preparation
# ==============================

# Import pandas
import pandas as pd

# ------------------------------
# STEP 1: Load Dataset
# ------------------------------

df = pd.read_excel("Dataset for Data Analytics.xlsx")

print("DATASET LOADED SUCCESSFULLY\n")

# ------------------------------
# STEP 2: Display First 5 Rows
# ------------------------------

print("FIRST 5 ROWS OF DATASET:\n")
print(df.head())

print("\n==============================\n")

# ------------------------------
# STEP 3: Dataset Information
# ------------------------------

print("DATASET INFORMATION:\n")
print(df.info())

print("\n==============================\n")

# ------------------------------
# STEP 4: Statistical Summary
# ------------------------------

print("STATISTICAL SUMMARY:\n")
print(df.describe())

print("\n==============================\n")

# ------------------------------
# STEP 5: Clean Column Names
# ------------------------------

df.columns = df.columns.str.strip()

print("COLUMN NAMES CLEANED\n")

# ------------------------------
# STEP 6: Check Missing Values
# ------------------------------

print("MISSING VALUES IN EACH COLUMN:\n")
print(df.isnull().sum())

print("\n==============================\n")

# ------------------------------
# STEP 7: Handle Missing Values
# ------------------------------

# Fill missing CouponCode values
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

print("MISSING VALUES HANDLED\n")

# ------------------------------
# STEP 8: Remove Duplicate Rows
# ------------------------------

before_duplicates = df.shape[0]

df = df.drop_duplicates()

after_duplicates = df.shape[0]

print("DUPLICATE ROWS REMOVED")
print("ROWS BEFORE:", before_duplicates)
print("ROWS AFTER :", after_duplicates)

print("\n==============================\n")

# ------------------------------
# STEP 9: Verify Duplicate Order IDs
# ------------------------------

duplicate_ids = df['OrderID'].duplicated().sum()

print("DUPLICATE ORDER IDs:")
print(duplicate_ids)

print("\n==============================\n")

# ------------------------------
# STEP 10: Standardize Date Format
# ------------------------------

df['Date'] = pd.to_datetime(df['Date'])

df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

print("DATE FORMAT STANDARDIZED\n")

# ------------------------------
# STEP 11: Standardize Text Columns
# ------------------------------

df['PaymentMethod'] = df['PaymentMethod'].str.title()

df['OrderStatus'] = df['OrderStatus'].str.title()

df['ReferralSource'] = df['ReferralSource'].str.title()

print("TEXT COLUMNS STANDARDIZED\n")

# ------------------------------
# STEP 12: Final Dataset Check
# ------------------------------

print("FINAL DATASET INFORMATION:\n")
print(df.info())

print("\n==============================\n")

print("FINAL FIRST 5 ROWS:\n")
print(df.head())

print("\n==============================\n")

# ------------------------------
# STEP 13: Save Cleaned Dataset
# ------------------------------

df.to_excel("Cleaned_Dataset.xlsx", index=False)

print("CLEANED DATASET SAVED SUCCESSFULLY!")

print("\nPROJECT COMPLETED SUCCESSFULLY!")
```
